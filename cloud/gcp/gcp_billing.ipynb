# GCP Billing

## Billing Resource Organization on GCP

### Cloud Resource Hierarchy
* Hierarchy of ownership binding resources to a single parent
* Establishes access and policy inheritance

Hierarchy Structure
* Organization (Company)
    * Folders (Departments, Teams, Products)
        * Projects (Dev GCP Project, Test GCP Project, Prod GCP Project)
            * Resources (Compute Engine Instance, App Engine Services)

### Billing Account Perspective
Under Org Layer
* Org policies are applied to the billing account
* The org can have multiple billing accounts

Relationship to Projects
* All projects must have a billing account
* A billing account can have multiple projects (one-to-many)

Account Management
* Billing account access control applies to all included projects
* Cost management applies to all included projects

## Managing Billing Access

### Control Billing Access with IAM Roles
* By limiting who can create new billing accounts and who can link them to projects, you protect cost overruns
* <ins>IAM</ins>: Who (Member), can do what (Role), on which resource (Resource)

Control Access
* <ins>Billing Account Administrator</ins>>: owns the billing account. Manages and assigns access to others.
* <ins>Billing Account Creator</ins>>: can create new billing accounts in the org but cannot assign rights to others
* <ins>Billing Account User</ins>>: can associate projects with billing accounts. Often paired with the "Project Creator" role

IAM Access Scopes
* Roles can be assigned at different layers. Roles assigned at organization layer are inherited by all attached billing accounts
* Best practices
    * Follow the principle of least privilege and assign only the necessary permissions to perform the role
    * Assign multiple billing account admins to ensure operation BCP (if one admin is unavailable, another can manage)


## Cost Visibility & Analysis

### GCP Billing Reports
* Provides visibility into cloud costs

### Export Billing Data to BigQuery
* Able to get a more fine-grained view into billing data in BigQuery
* You need the Billing Account Administrator role to configure a billing export to BigQuery
* Will only include data from after the export was setup 

### Billing Alerts
* Can be setup in the Billing service to automatically alert you

### Tools to Reduce Costs
* Rightsizing VMs - automatically generated recommendations to increase/decrease VMs 
* Sustained Use Discounts - automatic feature that gives you discounts
* Committed Use Discounts - billed monthly for 1-3 years of committment. Billed based on compute, not specific machines
* Preemptible VMs - save up to 80%. Max life of 24 hours. Good for fault tolerant, batch compute jobs
* Cloud Storage Classes - Standard, Nearline, Coldline, Archive
* Cloud Storage Lifecycle Policies - Automatically delete/change class of old data
* Free Tier Resources - low usage of many GCP services is always free.
